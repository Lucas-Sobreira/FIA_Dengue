## Lendo URL

```python
from urllib.request import urlopen
import json

url = 'https://olinda.bcb.gov.br/olinda/servico/Informes_Agencias/versao/v1/odata/Agencias?$top=100&$format=json'

response = urlopen(url)
data_json = json.loads(response.read())
print(data_json)
```

```python
import pandas as pd 

pd.set_option('max_colwidth', -1)
```

### Consultando apenas 1 Municipio

```python
url = 'https://info.dengue.mat.br/api/alertcity'

search_filter = (
    'geocode=3304557&disease=dengue&format=csv&' +
    'ew_start=1&ew_end=50&ey_start=2017&ey_end=2017'
)

df = pd.read_csv('%s?%s' % (url, search_filter))
print(url + search_filter)
df.head()
```

### Consultando Municipios da Bahia

```python
df_bahia = pd.read_excel('municipios.xlsx', sheet_name='Bahia')
df_bahia.head()
```

```python
for i in df_bahia.head(5)['Codigo']: 
    print(i)
```

```python
url = 'https://info.dengue.mat.br/api/alertcity'

df_geral = pd.DataFrame({})

for i in df_bahia.head(5)['Codigo']: 
    search_filter = (
        'geocode={}&disease=dengue&format=csv&'.format(i) +
        'ew_start=1&ew_end=50&ey_start=2017&ey_end=2017'
    )

    df = pd.read_csv('%s?%s' % (url, search_filter))
    df_geral = df_geral.append(df)
    print(df_geral.shape)
    # print(url + search_filter)
```
