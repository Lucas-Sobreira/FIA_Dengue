## Dengue

```python
import pandas as pd 
from tqdm import tqdm

pd.set_option('max_colwidth', -1)
```

### Consultando todos os Municipios

```python
df_municipios = pd.read_excel('municipios.xlsx', sheet_name='Sheet_01')
print('Tamanho do Dataframe: {}'.format(df_municipios.shape))
df_municipios.head()
```

```python
url = 'https://info.dengue.mat.br/api/alertcity'

df_dengue_ml = pd.DataFrame({})

# sem_end=52
year = 2021

for disease in ['dengue', 'chikungunya', 'zika']:
    print(year, disease)
    for cod_mun in tqdm(df_municipios['Código']): 
        search_filter = (
            'geocode={}&disease={}&format=csv&'.format(cod_mun, disease) +
            'ew_start=1&ew_end={}&ey_start={}&ey_end={}'.format(52, year, year)
        )
        
        df = pd.read_csv('%s?%s' % (url, search_filter))
        df['disease'] = disease
        df['year'] = year
        df_dengue_ml = df_dengue_ml.append(df)
        # print(df_dengue_ml.shape, "|" , year, "|" , disease)
        # print(url + search_filter)

print(df_dengue_ml.shape)

df_dengue_ml.head()
```
